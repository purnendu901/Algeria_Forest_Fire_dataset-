Algeria Forest Fire Dataset

About Dataset


The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attributes and 1 output attribute (class)
The 244 instances have been classified into fire (138 classes) and not fire (106 claes) c

lasses.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python


dataset=pd.read_csv('Algerian_forest_fires_dataset.csv' )
```

```python
dataset.head()
```

```python
dataset.info()
```

## DATA Cleaning

```python
dataset[dataset.isnull().any(axis=1)]
```

## Converted to 2 set based on Region  from 122th index, we can make a new column based on the redion:

1. Bejaia Region
2. Sidi-Bel REgion

add new column with region

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1

```

```python
dataset.info()
```

```python
dataset[["Region"]]=dataset[["Region"]].astype(int)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
##re,move all null values
dataset= dataset.dropna().reset_index(drop=True)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.iloc[[122]]
```

```python
##remove 122nd row
dataset=dataset.drop(122).reset_index(drop=True)
```

```python
##fix the spaced in column names
dataset.columns=dataset.columns.str.strip()
```

```python
dataset.columns
```

##changing types to float and integer


```python
dataset.columns
```

```python
dataset[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=dataset[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
dataset.info()
```

```python
objects=[features for features  in dataset.columns if dataset[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        dataset[i]=dataset[i].astype(float)
```

```python

dataset[['Classes']]=dataset[['Classes']].astype(object)
```

```python
dataset.info()
```

```python
##save the cleaned dataset
dataset.to_csv('Algerian_forest_fires_cleaned_dataset',index=False)
```

```python
dataset['Classes']=np.where(dataset['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
## Monthly Fire Analysis
datasettemp=dataset.loc[dataset['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dataset)
plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title("Fire Analysis of Sidi- Bel Regions",weight='bold')
```

```python
## Monthly Fire Analysis
datasttemp=dataset.loc[dataset['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dataset)
plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title("Fire Analysis of Bejaia Regions",weight='bold')
```

Its observed that August and September had the most number of forest fires for both regions. And from the above plot of months, we can understand few things

Most of the fires happened in August and very high Fires happened in only 3 months - June, July and August.
